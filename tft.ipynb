```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.impute import SimpleImputer

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, BatchNormalization, Dropout, ReLU, Flatten

# Define a custom wrapper function
class CustomKerasRegressor:
    def __init__(self, build_fn, **kwargs):
        self.build_fn = build_fn
        self.kwargs = kwargs
        self.model = None
    
    def fit(self, X, y, **fit_kwargs):
        self.model = self.build_fn(**self.kwargs)
        self.model.fit(X, y, **fit_kwargs)
        return self
    
    def predict(self, X):
        return self.model.predict(X)

# Load the data
data = pd.read_csv('two_year_with_holiday_2024Fall.csv')

# Convert 'Date/Time' to datetime object
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Sort data by 'Date/Time'
data = data.sort_values('Date/Time')

# Set 'Date/Time' as the index
data.set_index('Date/Time', inplace=True)

# Handle missing values (choose one of the methods)
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

data['Traffic Volume'] = imp_mean.fit_transform(data['Traffic Volume'].values.reshape(-1, 1))

# Extract hour, day of week, and month
data['Hour'] = data.index.hour
data['DayOfWeek'] = data.index.dayofweek
data['Month'] = data.index.month

# Optional: Drop 'Sensor' column if it's not useful
data.drop('Sensor', axis=1, inplace=True)

print(data.isnull().sum())
# Initialize the scaler
scaler = MinMaxScaler(feature_range=(0, 1))

# List of features to scale
features_to_scale = ['Traffic Volume', 'Holiday', 'Hour', 'DayOfWeek', 'Month']

# Scale the features
data_scaled = scaler.fit_transform(data[features_to_scale])

# Convert back to DataFrame
data_scaled = pd.DataFrame(data_scaled, columns=features_to_scale, index=data.index)

# Function to create sequences for RNN
def create_sequences(dataset, seq_length):
    X = []
    y = []
    for i in range(len(dataset) - seq_length):
        X.append(dataset.iloc[i:i+seq_length].values)
        y.append(dataset.iloc[i+seq_length]['Traffic Volume'])
    return np.array(X), np.array(y)

# Define sequence length (e.g., past 24 hours)
sequence_length = 24

# Create sequences
X, y = create_sequences(data_scaled, sequence_length)

# Display shapes
print(f"Feature shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Calculate the split index
train_size = int(0.8 * len(X))

# Split the data
X_train = X[:train_size]
X_test = X[train_size:]
y_train = y[:train_size]
y_test = y[train_size:]

# Display the shapes
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")

# Function to build the RNN model
def build_model(units=50, activation='relu', optimizer='adam'):
    model = Sequential()
    # First Conv1D layer
    input_shape = (X_train.shape[1], X_train.shape[2])

    model.add(Conv1D(filters=32, kernel_size=2, padding='same', input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(ReLU())
    model.add(MaxPooling1D(pool_size=2))
    
    # Second Conv1D layer
    model.add(Conv1D(filters=64, kernel_size=2, padding='same'))
    model.add(BatchNormalization())
    model.add(ReLU())
    model.add(MaxPooling1D(pool_size=2))
    
    # Third Conv1D layer
    model.add(Conv1D(filters=128, kernel_size=2, padding='same'))
    model.add(BatchNormalization())
    model.add(ReLU())
    model.add(MaxPooling1D(pool_size=2))
    
    # Flatten and fully connected layers
    model.add(Flatten())
    model.add(Dense(50, activation='relu'))  # Fully connected layer
    model.add(Dense(1))  # Output layer
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Use the custom wrapper for model training
#model = CustomKerasRegressor(build_model, units=50, activation='relu', optimizer='adam')
model = build_model()

# Fit the model
model.fit(X_train, y_train, epochs=10, batch_size=16)

model.summary()
# Make predictions
y_pred = model.predict(X_test)

# Reshape and inverse transform to get original values
y_pred = y_pred.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

y_pred_inverse = scaler.inverse_transform(
    np.concatenate((y_pred, X_test[:, -1, 1:]), axis=1)
)[:, 0]
y_test_inverse = scaler.inverse_transform(
    np.concatenate((y_test, X_test[:, -1, 1:]), axis=1)
)[:, 0]

# Calculate Mean Squared Error
mse = mean_squared_error(y_test_inverse, y_pred_inverse)
print(f"Test MSE: {mse}")

# Calculate Mean Absolute Percentage Error and Accuracy
mape = mean_absolute_percentage_error(y_test_inverse, y_pred_inverse)
accuracy = 100 - (mape * 100)
print(f"Model Accuracy: {accuracy:.2f}%")

# Plot results
plt.figure(figsize=(14, 7))
plt.plot(y_test_inverse, label='Actual Traffic Volume')
plt.plot(y_pred_inverse, label='Predicted Traffic Volume')
plt.xlabel('Time Steps')
plt.ylabel('Traffic Volume')
plt.title('Actual vs Predicted Traffic Volume')
plt.legend()
plt.show()
```
