```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.impute import SimpleImputer

import tensorflow as tf
from tensorflow.keras.layers import Dense, LayerNormalization, Dropout, MultiHeadAttention, Concatenate, LSTM, Add


# Define a custom wrapper function
class CustomKerasRegressor:
    def __init__(self, build_fn, **kwargs):
        self.build_fn = build_fn
        self.kwargs = kwargs
        self.model = None
    
    def fit(self, X, y, **fit_kwargs):
        self.model = self.build_fn(**self.kwargs)
        self.model.fit(X, y, **fit_kwargs)
        return self
    
    def predict(self, X):
        return self.model.predict(X)

data = pd.read_csv('two_year_with_holiday_2024Fall.csv')

data['Date/Time'] = pd.to_datetime(data['Date/Time'])

data = data.sort_values('Date/Time')

data.set_index('Date/Time', inplace=True)

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

data['Traffic Volume'] = imp_mean.fit_transform(data['Traffic Volume'].values.reshape(-1, 1))

data['Hour'] = data.index.hour
data['DayOfWeek'] = data.index.dayofweek
data['Month'] = data.index.month

data.drop('Sensor', axis=1, inplace=True)

data['IsWeekend'] = data['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)

# i gpted this, shows periodic values (24hrs)
data['Hour_sin'] = np.sin(2 * np.pi * data['Hour'] / 24)
data['Hour_cos'] = np.cos(2 * np.pi * data['Hour'] / 24)
data['Month_sin'] = np.sin(2 * np.pi * data['Month'] / 12)
data['Month_cos'] = np.cos(2 * np.pi * data['Month'] / 12)

data['TrafficVolume_Lag1'] = data['Traffic Volume'].shift(1)
data['TrafficVolume_Lag24'] = data['Traffic Volume'].shift(24)
data['TrafficVolume_RollingMean3'] = data['Traffic Volume'].rolling(window=3).mean()
data['TrafficVolume_RollingStd3'] = data['Traffic Volume'].rolling(window=3).std()

data.fillna(method='bfill', inplace=True)

data.drop(['Hour', 'DayOfWeek', 'Month'], axis=1, inplace=True)

print(data.isnull().sum())

scaler = MinMaxScaler(feature_range=(0, 1))


features_to_scale = [
    'Traffic Volume', 'Holiday', 'IsWeekend', 
    'Hour_sin', 'Hour_cos', 'Month_sin', 'Month_cos', 
    'TrafficVolume_Lag1', 'TrafficVolume_Lag24', 
    'TrafficVolume_RollingMean3', 'TrafficVolume_RollingStd3'
]


data_scaled = scaler.fit_transform(data[features_to_scale])

data_scaled = pd.DataFrame(data_scaled, columns=features_to_scale, index=data.index)

def create_sequences(dataset, seq_length):
    X = []
    y = []
    for i in range(len(dataset) - seq_length):
        X.append(dataset.iloc[i:i+seq_length].values)
        y.append(dataset.iloc[i+seq_length - 1]['Traffic Volume'])  # Only the last value
    return np.array(X), np.array(y)

sequence_length = 24


X, y = create_sequences(data_scaled, sequence_length)


print(f"Feature shape: {X.shape}")
print(f"Target shape: {y.shape}")


train_size = int(0.8 * len(X))


X_train = X[:train_size]
X_test = X[train_size:]
y_train = y[:train_size]
y_test = y[train_size:]


print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")


def build_tft_model(seq_len, n_features, d_model=64, num_heads=4, dropout_rate=0.1):
    inputs = tf.keras.layers.Input(shape=(seq_len, n_features))

    lstm_out = LSTM(d_model, return_sequences=True)(inputs)
    lstm_out = Dropout(dropout_rate)(lstm_out)
    

    attention_out = MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(lstm_out, lstm_out)
    attention_out = Dropout(dropout_rate)(attention_out)
    attention_out = LayerNormalization(epsilon=1e-6)(attention_out)
    

    attention_out = Add()([lstm_out, attention_out])

    lstm_out_2 = LSTM(d_model, return_sequences=True)(attention_out)
    lstm_out_2 = Dropout(dropout_rate)(lstm_out_2)
    
    ffn_out = Dense(d_model, activation="relu")(lstm_out_2)
    ffn_out = Dropout(dropout_rate)(ffn_out)
    
    output = Dense(1)(ffn_out[:, -1, :])
    
    model = tf.keras.Model(inputs=inputs, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss="mean_squared_error")
    return model

X_train_tft = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2]))
X_test_tft = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2]))

tft_model = build_tft_model(seq_len=sequence_length, n_features=X_train.shape[2])
tft_model.fit(X_train_tft, y_train, epochs=10, batch_size=16)

tft_model.summary()

y_pred = tft_model.predict(X_test_tft)

y_pred = y_pred.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)


sequence_length = X_test.shape[1]  # This should be 24 based on setup


print("Adjusted y_pred_final shape:", y_pred.shape)
print("X_test[:, -1, 1:] shape:", X_test[:, -1, 1:].shape)


y_pred_inverse = scaler.inverse_transform(
    np.concatenate((y_pred, X_test[:, -1, 1:]), axis=1)
)[:, 0]
y_test_inverse = scaler.inverse_transform(
    np.concatenate((y_test, X_test[:, -1, 1:]), axis=1)
)[:, 0]


mse = mean_squared_error(y_test_inverse, y_pred_inverse)
print(f"Test MSE: {mse}")


mape = mean_absolute_percentage_error(y_test_inverse, y_pred_inverse)
accuracy = 100 - (mape * 100)
print(f"Model Accuracy: {accuracy:.2f}%")


plt.figure(figsize=(14, 7))
plt.plot(y_test_inverse, label='Actual Traffic Volume')
plt.plot(y_pred_inverse, label='Predicted Traffic Volume')
plt.xlabel('Time Steps')
plt.ylabel('Traffic Volume')
plt.title('Actual vs Predicted Traffic Volume')
plt.legend()
plt.show()
```
